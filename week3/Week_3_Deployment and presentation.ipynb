```python
# Install scikit-learn if not already installed
%pip install scikit-learn pandas

import pandas as pd
from sklearn.linear_model import LinearRegression

# Create a sample DataFrame
sample = pd.DataFrame({'feature1': [1.5], 'feature2': [3.2]})

# Create and train a simple model for demonstration
X_train = pd.DataFrame({'feature1': [1, 2, 3], 'feature2': [2, 3, 4]})
y_train = [10, 15, 20]
model = LinearRegression()
model.fit(X_train, y_train)

# Make prediction
prediction = model.predict(sample)
print(f"Predicted PM2.5: {prediction[0]:.2f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv('../Week_1_(Dataset_Preparation)/data/clean_pollution_data.csv')

# Convert numeric months to labels (if needed)
month_map = {
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
}
df['month'] = df['month'].map(month_map)

# Group and plot
monthly_avg = df.groupby('month')[['TEMP']].mean().round(2)

monthly_avg.plot(kind='bar', figsize=(10, 6), color=['skyblue'])
plt.title('Monthly Average Temperature')
plt.ylabel('Temperature (Â°C)')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Only 'TEMP' column exists for temperature in df
hourly_avg = df.groupby('hour')[['TEMP']].mean().round(2)

hourly_avg.plot(figsize=(10, 6), marker='o')
plt.title('Hourly Average Temperature Trend')
plt.xlabel('Hour of Day')
plt.ylabel('Temperature (Â°C)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['pm2.5'], bins=30, color='salmon', edgecolor='black')
plt.title('PM2.5 Pollution Level Distribution')
plt.xlabel('PM2.5 Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by hour and calculate average PM2.5
pm_by_hour = df.groupby('hour')['pm2.5'].mean().round(2)

# Plot
plt.figure(figsize=(10, 6))
pm_by_hour.plot(kind='line', marker='o', color='tomato')
plt.title('Average PM2.5 Levels by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('PM2.5')
plt.grid(True)
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```
